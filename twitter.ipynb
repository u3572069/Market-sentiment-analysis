```python

#like-weighted average
import pandas as pd
import glob
import operator
from datetime import datetime, timedelta
coin = "NFT"
import numpy as np
end = 1577923201+24*3600*1100
senti = []
length = []
for i in range(1577923201, end, 24*3600):
    print(glob.glob(coin+"_"+str(i)+".csv")[0])
    df1 = pd.read_csv(glob.glob(coin+"_"+str(i)+".csv")[0])
    length.append(df1.shape[0])
    try:
        df1["Text"] = df1["Text"].astype(str)
        # df1["Text"] = df1["Text"].apply(lambda x : ' '.join([s for s in x.split(' ') if s.find('@') == -1 and s.find('www') == -1 and s.find('https') == -1]))
    except:
        pass
    df1 = df1.replace(r'\W+', ' ', regex=True)
    l = []
    for rows in df1.itertuples():
        l.append(sentiment_scores(rows.Text))
    l9 = df1['likes'].to_list()
    mean_score = list(map(operator.mul, l, l9 ))
    try:
        senti.append({'Date': i, 'Sentiment': sum(mean_score)/sum(l9), 'Ticker': coin})
    except:
        senti.append({'Date': i, 'Sentiment': 0, 'Ticker': coin})

    
```
